```python

#11)answer1 has 1 code line more than answer 2 but takes less steps than answer2
for i in range(1,101): # 401 steps
    if i%3==0:
        if i%5==0:
            print("FizzBuzz")
        else:
            print("Fizz")
    elif i%5==0:
        print("Buzz")
    else:
        print(i)
```

Write a Python function that takes two strings as input and returns True if they are anagrams, False otherwise. Anagrams are words or phrases formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once. For example, “listen” and “silent” are anagrams

```python
#12) using for and if else
def is_anagrams(str1,str2):
    if len(str1)==len(str2):
        for letter in str1:
            if letter in str2:
                if letter==str1[len(str1)-1]:
                    for i in range(len(str1)):
                        if str1[i]==str2[i]:
                            if i==len(str1)-1:
                                print("They are same words,aligned same,enter diff words and check are they anagrams or not")
                                break
                            else:
                                continue
                        else:
                            print("Anagrams")
                            break
            else:
                print("Not Anagrams")
                break
    else:
        print("Not Anagrams")
str1=input("enter a string 1: ")
str1=str1.lower()
str2=input("enter a string 2: ")
str2=str2.lower()
is_anagrams(str1,str2)
```

```python
#12)using while and if else
def is_anagrams(str1,str2):
    if len(str1)==len(str2):
        i=0
        while i<len(str1):
            if str1[i]==str2[i]:
                i+=1
                if i==len(str1)-1:
                    print("They are same words,aligned same")
                    break
                else:
                    continue
            else:
                for j in range(i,len(str1)):
                    if str1[j] in str2:
                        if j==len(str1)-1:
                            print("Anagrams")
                        else:
                            continue
                    else:
                        print("Not Anagrams")
                        break
                break    
    else:
        print("Not Anagrams")
str1=input("enter a string 1: ")
str1=str1.lower()
str2=input("enter a string 2: ")
str2=str2.lower()
is_anagrams(str1,str2)
```
